```python
# We need to first single out all the values which are incorrect, we can do that
# By simply checking which ones are correct, and then negating the return.

def is_float(x):
    try:
        float(x)
    except:
        return False
    return True

# This snippet checks to see if a given value is a float. If it is then its true
# if not, then it goes into the except section and returns false.
```

```python
"""
This snippet selects the total_sqft and applies the is float function to it.

It then uses .head() to present the first 5.

 # df3[df3['total_sqft'].apply(is_float)].head()
 
When ran without a negate sign (~) , it will show all the positive results.

"""
df3[~df3['total_sqft'].apply(is_float)].head(10)

"""
When ran with the negate sign, it shows all the results that returned False
"""
```

```python
"""
To reduce the range variables to a single integer, we can take the average
of the variables that are given in a range format.

This function takes in the value of total_sqft and splits the string, each 
word in the string will add to an array. 

If the array size equals 2, take both of the values and calculate the average
value, that average now replaces the split value.
"""
def convert_sqrt_to_avg (x):
    tokens = x.split("-")
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1])) / 2
    try:
        return float(x)
    except:
        return None
```

```python
df4 = df3.copy()
df4["total_sqft"] = df4["total_sqft"].apply(convert_sqrt_to_avg)
```

## Part Four - Outlier Removal

Outliers need to be removed so the final data calculations do not become askew based on incorrect / false data.

To calculate outliers, you can get data from your data manager or provider and ask what would typically be in the realm of possibility, for example having 2 bedrooms in a 50 sqft apartment is not likely. S you can use this domain knowledge to quickly take out outliers:

```python
df5[df5.total_sqft / df5.bhk<300].head()
```

You can show these data points to your domain manager, he/she will also agree that it is incredibly unlikely for a 8 bedroom home to fit inside a 0 sqft radius.

```python
df5.shape
```

```python
"""
When making changes it is easier to create ver a new variable instead.
"""

df6 = df5[~(df5.total_sqft / df5.bhk<300)]
df6.shape
```

You can see some outliers got removed, this was a specific search done using domain knowledge however.

Another way of finding the possible outliers is using the describe function:

```python
df6.price_per_sqft.describe()
```

Here you can see that someone is charging 267 lakh per square foot which is very unlikely and also 176470 Lakh per square foot.

To filter out outliers such as this one, you can use <font color=orange> **Standard Deviation** </font> to filter out any values that are far out the range of the <font color=orange>**Standard Deviation**</font>. 

Remember, each location will have a different average price. Richer areas will not be around the same price per square foot as poorer areas. For this reason you must apply a function to individual locations.

We will achieve this by taking the mean of every location and the standard deviation and comparing them to isolate any outliers.

```python
"""

The parameter is of a DataFrame, this is a data structure that holds all the
locations.

"""

def remove_pps_outliers(df):
   
    # New dataframe without outliers.
    df_out = pd.DataFrame()
    
    # For each key (location), 
    # Create a sub-dataframe called subdf
    for key, subdf in df.groupby('location'):
        
        # calculate Mean using np (Numpy)
        m = np.mean(subdf.price_per_sqft)
        
        # Calculate Standard Deviation using np (Numpy)
        st = np.std(subdf.price_per_sqft)
        
        # Create a dataframe for newly calculated values.
        # Calculate the range and ensure the price_per_sqft is not outside.
        reduced_df = subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft <= (m+st))]
        
        # pd.concat means add together (x-axis , y-axis) 
        # ignore_index=False means start from zero and n -1.  
        df_out = pd.concat ([df_out , reduced_df] , ignore_index=True)
    return df_out

print(df6.shape)
df7 = remove_pps_outliers(df6)
df7.shape
df7.head()
```

**We should also remove properties where obvious inconsistencies exist for example a 3 bedroom house which will have a bigger sqft being cheaper than a 2 bedroom one**

To achieve this, you can create a dictionary that holds describe() data such as mean and count of houses for every number of bedrooms. This way any house can be referenced back to that general describe function and then be filtered if it beyond a threshold.

After cross referencing a house stats with these general stats, you can remove outliers such as properties who have a lower price than a 2 bedroom house but are a 3 bedroom house.

```python
def remove_bhk_outliers (df):
    
    # A place to hold all removed elements
    removed_outliers = np.array([])
    
    # Go into every location and group them by their location
    for location , location_df in df.groupby('location'):
        bhk_stats = {}
        
        # Within every location group them by the BHK
        for bhk , bhk_df in location_df.groupby ('bhk'):
            
            # Curly brackets starts a dictionary, gp into bhk_stats and apply to all BHK
            bhk_stats [bhk] = {
                'mean' : np.mean (bhk_df.price_per_sqft),
                'std' : np.std (bhk_df.price_per_sqft),
                'count' : bhk_df.shape[0]
            } # This bracket purpose is to gather these information about each 
              # location and BHK in that location.
        
        for bhk , bhk_df in location_df.groupby('bhk'):
            
            # Retrieve the stats of the BHK below the selected.
            stats = bhk_stats.get(bhk-1)
            
            # If stats has a value and is above 5 
            if stats and stats['count'] > 5 :
                
                # add the removed outliers and all price per sqft houses that are less than the mean
                removed_outliers = np.append(removed_outliers , bhk_df[bhk_df.price_per_sqft < (stats['mean'])].index.values)

     # Add them all to the variable and drop them all.                                        
    return df.drop (removed_outliers , axis='index')        
    
df8 = remove_bhk_outliers(df7)
df8.shape                                             
```

## Part Five - Model Building

First of all, machine learning algorithms find it difficult to interpret strings/text, therefore they need to be changed into numeric columns.

We can do this by using <font color = orange> **One-hot Encoding** or **"Dummies"** </font>, this is a Pandas method that simply provides each string a <font color=orange > **binary key** </font> to identify each string, such as below:

```python
df10.head()
```

```python
pd.get_dummies(df10.location)
```

Each location has now been given different values on a binary scale, this is a pretty straight forward method to change categoricals to numerical data.

```python
dummies = pd.get_dummies(df10.location)
dummies.head(3)
```

Now we want to add this onto the original dataframe from the previous parts so the information in those are also accessible and resonate with the correct locations, we want to avoid mixing up all the data.

```python
"""
Firstly, pd.concat() is used when two arrays need to be merged, they take an array in the 
parameter and you need to specify if you want the rows merged or columns.

Secondly, The final row is being dropped to avoid a dummies error that occurs called "Dummy 
Variable Trap", this trap occurs when an algorithm goes through all the binary numbers and does
not find a match, there is no default so therefore it fails, removing one of the columns provides
a default for it to return too if there is no match.

For example, imagine binary values for asians , hispanics , whites and african races. If a binary 
strip runs and concludes it is non of these races, it will stop, whereas by removing african races,
column, if it is not asian , hispanic or white, it must default to african race.

Similar to a try {} except{} function.

The drop function also needs to be told what axis the location is to be found in, in this case:
'columns'
"""

df11 = pd.concat([df10 , dummies.drop('others' , axis='columns')] , axis='columns')
df11.head(10)
```

```python
# Remember Ml algorithms don't understand text so Location needs to be dropped.
df12 = df11.drop('location' , axis='columns')
df12.head(3)
```

Now we are ready to start working on a machine learning algorithm:

```python
# It is a good habit to drop independant variables into an X or Y variable, remember these are 
# Only variables that are not contingent upon things.
x = df12.drop('price' , axis='columns')
x.head(5)

# Price is a dependant variable on price per sqft.
```

```python
# Still need access to the variable for future use so place it within another variable:
y = df12.price
y.head()
```

```python
"""
In ML, it is good practise to make two datasets with which you will work with:

1) Training Test Dataset - We use this for Model Training.
2) Test Dataset - To evaluate Test Performance.

Below is how to import a sample for training and testing.

test_size indicates 20% of the test sample will be a test dataset, the rest will be for training.
"""

from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
# Once the model it trained
lr_clf.fit(x_train , y_train)
# This will score how good the model is.
lr_clf.score (x_test , y_test)
```

Currently it outputs a response of 84% which is pretty good. Often Data Scientists will try numerous parameters to find and configure the highest possible result.

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

# ShuffleSplit will randomise the sample so all four testing have equal distribution of the datasample.
cv = ShuffleSplit (
    # Number of re-shuffling iterations
    n_splits=5 , 
    # Represents the size of the proportion thats being tested.
    test_size=0.2 ,
    # The seed used by the random number generator
    random_state=0
)

cross_val_score(LinearRegression(), x , y , cv=cv)
```

The above numbers are displaying the responses of the data using numerous different models. <font color=orange> **This is called K-Fold Cross Validation.** </font> As a data scientists, you should try to use the highest performing model you can, so you should try a variety of algorithms to get the best score.

SKLearn also has its own API that can run though a variety of algorithms and try to find the best one, this is called the <font color=orange> **GridSearchCV** </font>. It will run your dataset on different parameters and regressors.

Each algorithm has different parameters needed for it to run, 

Below is an example

```python
from sklearn.model_selection import GridSearchCV

# Import different algorithms.
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

# Now write a function

def find_best_gridsearch (x , y):
    algos = {
        # Provide all parameters for Linear Regression 
        'linear regression' : {
            'model' : LinearRegression() ,
            'params' : {
                'normalize' : [True , False]
            }
        },
        
        # Provide all parameters for Lasso
        'lasso' : {
            'model' : Lasso() , 
            'params' : {
                'alpha' : [1 , 2] , 
                'selection' : ['random' , 'cyclic']
            }
        },
        
        # Provide all parameters for Decision Tree
        'decision_tree' : {
            'model' : DecisionTreeRegressor() , 
            'params' : {
                'criterion' : ['mse' , 'friedman_mse'] , 
                'splitter' : ['best' , 'random']
            }
        }
    }
    
    # Keep hold of the scores
    scores = []
    
    # n_splits = 5 means split the data into 5 sections and apply the test on all 5 random sections.
    cv = ShuffleSplit (
        n_splits=5 , 
        test_size=0.2 ,
        random_state=0
    )
    for algo_name, config in algos.items():
        
        # Input the algorithm name and all of its parameters & the amount of cross validations.
        gs = GridSearchCV(config['model'] , config['params'] , cv=cv , return_train_score=False)
        
        # Input the dataset and dependant variables
        gs.fit(x , y)
        
        # Add all the calculations to scores
        scores.append ({
            'model' : algo_name, 
            
            # Calculate best score from the iterations
            'best_score' : gs.best_score_ , 
            
            # Calculate the best Parameters from the iterations
            'best_params' : gs.best_params_
        })
        
    return pd.DataFrame(scores , columns=['model' , 'best_score' , 'best_params'])


find_best_gridsearch(x,y)
```

As the output shows, the Linear Regression still provided the best output for the model.

```python
def predict_price (location , sqft , bath , bhk):
    loc_index = np.where(x.columns==location)[0][0]
    
    # np.zeros returns an array the size of the columns and fill them with zeros 
    X = np.zeros(len(x.columns))
    
    # first zero represents sqft
    X[0] = sqft
    X[1] = bath
    X[2] = bhk
    if loc_index >= 0: 
        X[loc_index] = 1
        
    return lr_clf.predict([X])[0]
```
